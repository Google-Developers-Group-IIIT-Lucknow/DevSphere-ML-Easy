```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
def prepare_data():
    # TODO: Define the dataset url
    # BUG: Where is the data coming from? url is not assigned
    
    # BUG: This fails because 'pd' isn't defined yet
    df = pd.read_csv('data.csv')
    
    print("Pre-cleaning null check:\n", df.isnull().sum())
    
    # TODO: The dataset has a null value. 
    # BUG: This line runs but doesn't actually update 'df'
    df.dropna(inplace=True) 
    
    return df
```

```python
def train_the_beast(df):
    # MISSION: Predict 'Scores' using 'Hours'
    # BUG: X and y are swapped. This is predicting Hours from Scores.
    y = df[['Scores']] 
    X = df[['Hours']] 
    
    # TODO: Initialize the LinearRegression model

    model = LinearRegression()
    model.fit(X, y) 
    
    # BUG: This will fail if 'model' isn't initialized
    
    # BUG: Is "train" the right scikit-learn method?
    # model.predict(y)
    return model, X, y
```

```python
def visualize_results(model, X, y):
    plt.figure(figsize=(8,5))
    # BUG: We want a scatter plot for the actual data, but someone used plot() for both!
    plt.scatter(X, y, color='blue', label='Actual Data')
    
    # BUG: Predicting using 'y' instead of 'X' creates a mess
    plt.plot(X, model.predict(X), color='red', linewidth=2, label='Regression Line')
    
    plt.title('Study Hours vs Percentage Score')
    plt.xlabel('Hours Studied')
    plt.ylabel('Percentage Score')
    plt.legend()
    plt.show()
```

```python
try:
    df_clean = prepare_data()
    trained_model, features, target = train_the_beast(df_clean)
    visualize_results(trained_model, features, target)
    print("✅ Level 1 Cleared! Your aura is increasing.")
except Exception as e:
    print(f"❌ Still Broken: {e}")
```
